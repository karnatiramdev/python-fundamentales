```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('countries (2).csv')
data
```

```python
#req:compare the papulation of us and china
```

```python
data.country=='United States'
```

```python
us=data[data.country=='United States']
```

```python
us
```

```python
plt.plot(us.year,us.population/10**6)
plt.show()
```

```python
China=data[data.country=='China']
```

```python
China
```

```python
plt.plot(China.year,China.population/10**6)
plt.show()
```

```python
plt.plot(us.year,us.population/10**6)
plt.plot(China.year,China.population/10**6)
plt.show()
```

```python
plt.plot(us.year,us.population/10**6)
plt.plot(China.year,China.population/10**6)
plt.title('comparing papulation of a countrys based on year')
plt.legend('China')
plt.legend('united states')
plt.xlabel('year')
plt.ylabel('papulation')
plt.show()
```
